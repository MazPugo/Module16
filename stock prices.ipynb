```python
#one plot with formatted date axis
#fig, ax = plt.subplots(figsize=(16, 9))
fig, (ax, ax2) = plt.subplots(nrows=2, sharex=True,figsize=(16, 9))
# Add x-axis and y-axis
ax.plot(df_kgh['Data'],df_kgh['Zamkniecie'],color='blue')
ax2.plot(df_miedz['Zamkniecie'],color='red')
#set facecolor on subplot
ax.set_facecolor('#A9B2FF')
ax2.set_facecolor('#A9B2FF')
# Set title and labels for axes
ax.set(xlabel="Date",ylabel="Zamkniecie(zl)",title="kgh",xlim=["02/01/2015", "30/06/2020"])
ax.set_ylabel("Zamkniecie(zl)",fontsize = 15) # Y label
ax2.set(xlabel="Date",ylabel="Zamkniecie(zl)",title="miedz",xlim=["02/01/2015", "30/06/2020"])
ax2.set_ylabel("Zamkniecie(zl)",fontsize = 15) # Y label
# Define the date format
date_form = DateFormatter("%Y")
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.YearLocator(1,day=2,month=1))
plt.setp(ax.get_xticklabels('Zamkniecie'), fontsize=12, fontweight="bold") 
ax.grid(color='w', linestyle='--')
ax2.grid(color='w', linestyle='--')
plt.show()
```
